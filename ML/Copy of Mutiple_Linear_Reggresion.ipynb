#Computing

```
max= [df[c].max() for c in df.columns] 
min= [df[c].min() for c in df.columns]
```

```
i=0
for c in df.columns:
    while(i<len(df.columns)): 
        df[c]=(df[c]-min[i])/(max[i]-min[i])
        i=i+1
        break
```

#Normalize Data Set

```
arr = df.values
x_train=[]
y1=[]
y2=[]
a=df.shape
for i in range(a[0]):                      
    x_train.append((arr[i][:-2]).tolist())
    y2.append(arr[i][-1])
    y1.append(arr[i][-2])
```

```
m=np.ones((50,1))
b=np.matrix(x_train)
b=np.concatenate((m,b),axis=1)
d=b.T
e=np.linalg.inv(np.matmul(d,b))
y1=np.matrix(y1)
y1=y1.T
y2=np.matrix(y2)
y2=y2.T
f=np.matmul(e,d)
c1=np.matmul(f,y1)
c2=np.matmul(f,y2)
```
