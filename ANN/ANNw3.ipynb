```
import tensorflow as tf
import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split
```

```
df = pd.read_csv('Iris.csv')
df
```

```
features = df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']].values
label = df['Species'].values.reshape(-1,1)
```

#Data Splitting

```
x_train, x_test, y_train, y_test = train_test_split(features, label, test_size=0.2)
```

```
encoder = OneHotEncoder()
encoder.fit(y_train)
y_train = encoder.transform(y_train).toarray()
y_test = encoder.transform(y_test).toarray()
```

```
scaler = StandardScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```
input_tensor = tf.placeholder(tf.float32,name="input")
label_tensor = tf.placeholder(tf.float32,name="output")

parameters = {
    'W1' : tf.Variable(
        tf.random.normal((4,64)),
        dtype=tf.float32
    ),
    'B1' : tf.Variable(
        tf.random.normal((1,64)),
        dtype=tf.float32
    ),
    'W2' : tf.Variable(
        tf.random.normal((64,64)),
        dtype=tf.float32
    ),
    'B2' : tf.Variable(
        tf.random.normal((1,64)),
        dtype=tf.float32
    ),

    'W3' : tf.Variable(
        tf.random.normal((64,3)),
        dtype=tf.float32
    ),
    'B3' : tf.Variable(
        tf.random.normal((1,3)),
        dtype=tf.float32
    ),
}
```

```
def forward(x):
    z1 = tf.matmul(x,parameters['W1'])+ parameters['B1']
    a1 = tf.nn.sigmoid(z1)

    z2 = tf.matmul(a1,parameters['W2'])+ parameters['B2']
    a2 = tf.nn.sigmoid(z2)

    z3 = tf.matmul(a2,parameters['W3'])+ parameters['B3']
    a3 = tf.nn.sigmoid(z3)

    return a3
```

```
with tf.Session() as sess:
    logits_tensor = forward(input_tensor)
    #loss_tensor =(0.5 * (logits_tensor - label_tensor)**2).sum(axis=1).mean()
    loss_tensor = tf.reduce_mean(0.5*(logits_tensor - label_tensor)**2)
    optimizer = tf.train.AdamOptimizer(0.01).minimize(loss_tensor)

    sess.run(tf.global_variables_initializer())
    for epoch in range(500):
        sess.run(optimizer,
                 feed_dict={
                     input_tensor: x_train,
                     label_tensor: y_train
                 })
        
        print(
            sess.run(loss_tensor,
                    feed_dict={
                        input_tensor: x_train,
                        label_tensor: y_train
                    }
            )
        )
```
