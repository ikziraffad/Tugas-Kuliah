```
#Import Library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow.compat.v1 as tf
from sklearn.model_selection import train_test_split
import tensorflow as tf
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler,OneHotEncoder
```

```
#Import Dataset
data = pd.read_csv('classification.csv')
data
```

```
#Feature Selection

for i in range(len(data)):
  #Cap Shape
  if data['cap-shape'][i] == 's':
    data['cap-shape'][i] = 6
  elif data['cap-shape'][i] == 'k':
    data['cap-shape'][i] = 5
  elif data['cap-shape'][i] == 'f':
    data['cap-shape'][i] = 4
  elif data['cap-shape'][i] == 'x':
    data['cap-shape'][i] = 3
  elif data['cap-shape'][i] == 'c':
    data['cap-shape'][i] = 2
  elif data['cap-shape'][i] == 'b':
    data['cap-shape'][i] = 1
  
  #Odor
  if data['odor'][i] == 's':
    data['odor'][i] = 9
  elif data['odor'][i] == 'p':
    data['odor'][i] = 8
  elif data['odor'][i] == 'n':
    data['odor'][i] = 7
  elif data['odor'][i] == 'm':
    data['odor'][i] = 6
  elif data['odor'][i] == 'f':
    data['odor'][i] = 5
  elif data['odor'][i] == 'y':
    data['odor'][i] = 4
  elif data['odor'][i] == 'c':
    data['odor'][i] = 3
  elif data['odor'][i] == 'l':
    data['odor'][i] = 2
  elif data['odor'][i] == 'a':
    data['odor'][i] = 1

  #Habitat
  if data['habitat'][i] == 'd':
    data['habitat'][i] = 7
  elif data['habitat'][i] == 'w':
    data['habitat'][i] = 6
  elif data['habitat'][i] == 'u':
    data['habitat'][i] = 5
  elif data['habitat'][i] == 'p':
    data['habitat'][i] = 4
  elif data['habitat'][i] == 'm':
    data['habitat'][i] = 3
  elif data['habitat'][i] == 'l':
    data['habitat'][i] = 2
  elif data['habitat'][i] == 'g':
    data['habitat'][i] = 1


data
```

```
x_data = data[['cap-shape','cap-color','odor','stalk-color-above-ring','stalk-color-below-ring','veil-color','ring-number','habitat']]
y_data = data[['class']]
```

```
scaler = MinMaxScaler()
x_data = scaler.fit_transform(x_data)

encoder = OneHotEncoder(sparse=False)
y_data = encoder.fit_transform(y_data)

print(x_data)

pca = PCA(n_components=4)
pca = pca.fit(x_data)
x_data = pca.transform(x_data)
print(x_data)
```

```
layer = {
    'input' : 4,
    'hidden' : 3,
    'output' : 2
}

weight = {
    'to_hidden' : tf.Variable(tf.random.normal([layer['input'],layer['hidden']])),
    'to_output' : tf.Variable(tf.random.normal([layer['hidden'],layer['output']]))
}

bias = {
    'to_hidden' : tf.Variable(tf.random.normal([layer['hidden']])),
    'to_output' : tf.Variable(tf.random.normal([layer['output']]))
}
```

```
#Spliting data
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data,train_size=0.7)
```

```
#Spliting data for validation
x_validation, _, y_validation, _ = train_test_split(x_train, y_train, train_size=0.2)
```

```
x = tf.placeholder(tf.float32, [None, layer['input']])
target = tf.placeholder(tf.float32, [None,layer['output']])
```

```
def predict():
    wx_b1= tf.matmul(x, weight['to_hidden']) + bias['to_hidden']
    y1= tf.nn.sigmoid(wx_b1)

    wx_b2= tf.matmul(y1, weight['to_output']) + bias['to_output']
    y2= tf.nn.sigmoid(wx_b2)
    return y2
```

```
y= predict()
epoch = 2500
alpha = 0.02
error = tf.reduce_mean(0.5 * (target - y) **2)
optimizer = tf.train.GradientDescentOptimizer(alpha)
train = optimizer.minimize(error)
```

```
with tf.Session() as sess:
  saver = tf.train.Saver()
  sess.run(tf.global_variables_initializer())
  best_error = float('inf')
  for i in range(epoch):
    sess.run(
        train,
        feed_dict = {
            x: x_train,
            target: y_train
        }
    )

    if i % 25 == 0:
      current_error = sess.run(
          error,
          feed_dict = {
              x: x_train,
              target: y_train
          }
      )
      print(f'EPOCH : {i} | ERROR : {current_error} |')
    if i % 125 == 0:
      validation_error = sess.run(
          error,
          feed_dict = {
              x: x_validation,
              target: y_validation
          }
      )
      if validation_error < best_error:
        saver.save(sess, './checkpoint.ckpt')
      print(f'EPOCH : {i} | VALIDATION_ERROR : {validation_error} |')

  true_prediction = tf.equal(tf.argmax(y, axis = 1), tf.argmax(target, axis = 1))
  accuracy = tf.reduce_mean(tf.cast(true_prediction, tf.float32))
  saver.restore(sess, './checkpoint.ckpt')
  accuracy = sess.run(
      accuracy,
      feed_dict = {
          x: x_test,
          target: y_test
      }
  )  
  print(f'ACCURACY: {accuracy:}')
```
