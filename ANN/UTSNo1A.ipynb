```python
#Importing libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Importing dataset
from sklearn.datasets import load_digits
digits=load_digits(n_class=2) #Memakai n_class = 2 untuk hanyak memakai 2 kelas
```

```python
#Define variables
n_samples=len(digits.images)
print("Number of samples in the data set is :"+ str(n_samples))

x=digits.images.reshape((n_samples,-1)) # Reshaping the vector into the number of samples
print("Shape of input matrix x is : "+str(x.shape))

#Declaring y as the targets
y=digits.target
```

```python
#Split dataset for training and testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler #Normalize the data using standard scaler because we don't know the min max of the x
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
#Initialize Variables
weight = np.random.normal(size = (64, 1)) #Randomize weight with size (64,1) to fit the size of x
bias = np.random.normal(size = 1)
NUM_OF_EPOCH = 500
ALPHA = 0.5
```

```python
#Linear combination operation of input vectors and weight vectors
def feed_forward(inputs): #W.P + b
    value = np.matmul(inputs, weight)
    value += bias
    return activation_function(value) 

#Binary/hard limit Activation Function
def activation_function(x):
    if x >= 0:
        return 1
    else:
        return 0
```
