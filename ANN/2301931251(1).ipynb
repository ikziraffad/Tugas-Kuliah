```python
import tensorflow as tf
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('dataset.csv')
```

```python
features = df[['Time','Premise','Age','Gender']].values
label = df['Crime Code'].values.reshape(-1,1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(features, label, test_size=0.2)
```

```python
sc = StandardScaler()
sc.fit(x_train)
x_train = sc.transform(x_train)
x_test = sc.transform(x_test)

encoder = OneHotEncoder()
encoder.fit(y_train)
y_train = encoder.transform(y_train).toarray()
y_test = encoder.transform(y_test).toarray()
```

```python
input_tensor = tf.placeholder(tf.float32,name="input")
label_tensor = tf.placeholder(tf.float32,name="output")

parameters = {
    'W1' : tf.Variable(
        tf.random.normal((4,64)),
        dtype=tf.float32
    ),
    'B1' : tf.Variable(
        tf.random.normal((1,64)),
        dtype=tf.float32
    ),
    'W2' : tf.Variable(
        tf.random.normal((64,64)),
        dtype=tf.float32
    ),
    'B2' : tf.Variable(
        tf.random.normal((1,64)),
        dtype=tf.float32
    ),

    'W3' : tf.Variable(
        tf.random.normal((64,3)),
        dtype=tf.float32
    ),
    'B3' : tf.Variable(
        tf.random.normal((1,3)),
        dtype=tf.float32
    ),
}
```

```python
def forward():
    z1 = tf.matmul(input_tensor,parameters['W1'])+ parameters['B1']
    a1 = tf.nn.sigmoid(z1)

    z2 = tf.matmul(a1,parameters['W2'])+ parameters['B2']
    a2 = tf.nn.sigmoid(z2)

    z3 = tf.matmul(a2,parameters['W3'])+ parameters['B3']
    a3 = tf.nn.softmax(z3)

    return a3
```

```python
epoch = 5000
alpha = 0.2
y = forward()
error = tf.reduce_mean(0.5 * (label_tensor-y)**2)
train = tf.train.GradientDescentOptimizer(alpha).minimize(error)
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for i in range(epoch):
        sess.run(
            train,
            feed_dict={
                input_tensor: x_train,
                label_tensor: y_train
            })
        if i % 100 == 0:
            current_error = sess.run(
                error,
                feed_dict={
                    input_tensor: x_train,
                    label_tensor: y_train
                })
            true_prediction = tf.equal(tf.argmax(y,axis=1),tf.argmax(label_tensor,axis=1))
            accuracy = tf.reduce_mean(tf.cast(true_prediction,tf.float32))
            accuracy = sess.run(
                accuracy,
                feed_dict={
                    input_tensor: x_test,
                    label_tensor: y_test
                })
            print(f'EPOCH:{i+1} | ACCURACY: {accuracy*100}% | ERRROR: {current_error:.4f}')
```
