```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

#initialize the training patterns

```python
training = [[np.array([1, 1]).transpose(), 0], 
            [np.array([1, 0]).transpose(), 1],
            [np.array([0, 1]).transpose(), 1],
            [np.array([1, 0]).transpose(), 0]]
```

#initialize the variable

```python
error_grp = []
iter_grp = []
W = np.array([0.5, 0.3])
```

#LMS Function

```python
def lms(W):
    B= 0.1
    i = 1
    alpha = [0.5, 0.5]
    maxIter = 2
    E = 0
    while (i <= maxIter):
        E = 0
        for pair in training:
            U = np.dot(W.transpose(), pair[0]) + B
            W = W + np.dot((alpha[1]*(pair[1] - U)), pair[0])
            E = E + np.power((pair[1] - U), 2) * 0.5
            B = B + alpha[1]*(pair[1] - U)
            #print(U)
            #print(W)
            #print(E)
            #print(B)
        #print('Error: ' + str(E))
        #Put the error in the array after going thru the whole training pattern
        print('>epoch=%d, lrate=%.3f, error=%.3f' % (i, 0.5, E))
        error_grp.append(E)
        iter_grp.append(i)
        i = i + 1
    #print('i: ' + str(i))
    print('Final error: ' + str(E))
    print('final weight: ' + str(W))
```

```python
lms(W)
```

#Draw the graph

```python
#Draw the graph
plt.plot(iter_grp, error_grp)
plt.ylabel('Error')
plt.xlabel('No. of iterations')
plt.show()
#plt.savefig('0.01.png')
```
