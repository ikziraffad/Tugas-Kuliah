```python
import numpy as np
```

```python
weight = np.random.normal(size = (2, 1))
bias = np.random.normal(size = 1)

input_data = [
    [0 , 0],
    [0 , 1],
    [1 , 0],
    [1 , 1]
]

output_data = [0 , 1 , 1 , 1] #target
```

```python
def feed_forward(inputs): #W.P + b
    value = np.matmul(inputs, weight)
    value += bias
    return activation_function(value) 

#relu Activation Function
def activation_function(x):
    if x >= 0:
        return 1
    else:
        return 0
```

```python

NUM_OF_EPOCH = 100
ALPHA = 0.1
```

```python
for i in range(1, NUM_OF_EPOCH + 1):
    randIndex = np.random.randint(0, len(input_data))
    data_training = input_data[randIndex]

    output_training = feed_forward(data_training)
    error = output_data[randIndex] - output_training

    #update weight
    weight = np.add(weight, np.array(data_training).reshape(2, 1) * ALPHA * error)

    #update Bias
    bias = bias + (ALPHA * error)

    #testing
    if i % 20 == 0:
        corretct_result = 0
        for i , inp in enumerate(input_data):
            output_testing = feed_forward(inp)
            if output_testing == output_data[i]:
                corretct_result += 1
            
        print(f'Accuracy : {corretct_result/len(input_data)  * 100}%')
```
